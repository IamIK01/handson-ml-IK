### 2. Draw an ANN using the original artificial neurons that computes \(A $\oplus$ B\) (where \($\oplus$\) represents the XOR operation).

The XOR operation \(A $\oplus$ B\) can be decomposed as \(A $\oplus$ B = (A $\land$ $\neg$ B) $\lor$ ($\neg$ A $\land$ B)\). To construct an ANN for this:

- **Layer 1**:
  - Neuron 1: Computes \( $\neg$ A \) (NOT A)
  - Neuron 2: Computes \( $\neg$ B \) (NOT B)
- **Layer 2**:
  - Neuron 3: Computes \( A $\land$ \neg B \) (A AND NOT B)
  - Neuron 4: Computes \( $\neg$ A $\land$ B \) (NOT A AND B)
- **Layer 3**:
  - Neuron 5: Computes \( (A $\land$ $\neg$ B) $\lor$ ($\neg$ A $\land$ B) \) (OR of the outputs of Neuron 3 and Neuron 4)

**Visualization:**
```
   A ----> [¬] ----\
                     \
                      --> [AND] ---> \
   B ----> [¬] --------> [AND] ------> [OR] ---> A ⊕ B
                      /
   A ----------------/
                     /
   B ---------------/
```

### 3. Why is it generally preferable to use a Logistic Regression classifier rather than a classical Perceptron?

- **Flexibility**: Logistic Regression provides probabilities and is flexible under non-linear separability.
- **Output Interpretation**: It offers a probabilistic interpretation, suitable for real-world applications.
- **Robust Training**: It uses a loss minimization approach which is generally more effective and robust.

**Perceptron Adjustment**: To make a Perceptron equivalent to a Logistic Regression classifier, change the activation function from a step function to a logistic sigmoid function. This modification allows the Perceptron to output probabilities.

### 4. Why was the logistic activation function a key ingredient in training the first MLPs?

- **Non-linearity**: It introduces non-linearity, crucial for learning complex patterns.
- **Differentiability**: Its differentiability enables gradient-based optimization techniques like backpropagation.
- **Output Range**: The output range (0, 1) naturally fits probability outputs for binary classification.

### 5. Name three popular activation functions. Can you draw them?

Three popular activation functions include:

- **Sigmoid**: \( f(x) = $\frac{1}{1 + e^{-x}}$ \)
- **ReLU (Rectified Linear Unit)**: \( f(x) = $\max(0, x)$ \)
- **Tanh (Hyperbolic Tangent)**: \( f(x) = $\tanh(x)$ \)

**Python Plotting Code:**

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 1000)
sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

plt.figure(figsize=(12, 8))
plt.plot(x, sigmoid, label='Sigmoid')
plt.plot(x, relu, label='ReLU')
plt.plot(x, tanh, label='Tanh')
plt.title('Activation Functions')
plt.xlabel('Input value (x)')
plt.ylabel('Output value')
plt.legend()
plt.grid(True)
plt.show()
````

### 6. MLP with One Input, One Hidden Layer, and One Output Layer
- **Shape of the input matrix \($X$\)**:
  - \($X$\) will have a shape of \($[n, 10]$\), where \($n$\) is the number of instances in the batch.

- **Shapes of the hidden layer's weight vector \($W_h$\) and bias vector \($b_h$\)**:
  - \($W_h$\) will have a shape of \($[10, 50]$\) because there are 10 input neurons and 50 neurons in the hidden layer.
  - \($b_h$\) will have a shape of \($[50]$\) since there is one bias term for each neuron in the hidden layer.

- **Shapes of the output layer's weight vector \($W_o$\) and bias vector \($b_o$\)**:
  - \($W_o$\) will have a shape of \($[50, 3]$\) because there are 50 neurons in the hidden layer and 3 output neurons.
  - \($b_o$\) will have a shape of \($[3]$\), one for each output neuron.

- **Shape of the network’s output matrix \($Y$\)**:
  - \($Y$\) will have a shape of \($[n, 3]$\) since there are \($n$\) instances in the input and 3 output neurons.

- **Equation to compute the network's output matrix \($Y$\)**:

  \[
      $Y = \text{ReLU}(\text{ReLU}(X W_h + b_h) W_o + b_o)$
  \]

  
  This equation first computes the activations of the hidden layer, applies the ReLU activation function, then computes the output layer's inputs and applies the ReLU function again to get the final outputs.

### 8. Backpropagation and Reverse-Mode Autodiff
- **What is Backpropagation?**
  - Backpropagation is an algorithm used to compute the gradient of the loss function of a neural network with respect to its weights. It is used for training neural networks, where gradients are propagated back through the network to update the weights using gradient descent.

- **How does it work?**
  - Backpropagation calculates the gradient starting from the output layer back to the input layer. It efficiently computes the gradient of the loss function by using the chain rule from calculus, minimizing the loss by adjusting the weights iteratively.

- **Difference between Backpropagation and Reverse-Mode Autodiff:**
  - Backpropagation is a specific application of reverse-mode autodiff, used specifically in the training of neural networks. Reverse-mode autodiff is a broader technique used in automatic differentiation to calculate derivatives of functions efficiently, applicable in various contexts beyond neural networks.

### 9. Hyperparameters in a Basic MLP and Overfitting Adjustments
- **List of Hyperparameters**:
  - Number of layers and number of neurons in each layer.
  - Activation functions.
  - Weight initialization method.
  - Learning rate of the optimizer.
  - Regularization techniques (e.g., L1, L2, dropout).
  - Batch size.
  - Number of epochs.
  
- **Adjusting Hyperparameters to Prevent Overfitting**:
  - Increase regularization (e.g., higher L1/L2 regularization values, more dropout).
  - Reduce the number of neurons or layers (simplify the model).
  - Use early stopping.
  - Decrease the learning rate.
  - Increase the batch size.
  - Augment the training data.

These adjustments can help control overfitting by constraining the capacity of the model or by stopping training before the model learns the noise in the training data.

````python
def leaky_relu(x, alpha=0.01):
    """
    Leaky ReLU activation function.
    :param x: Input data.
    :param alpha: Slope for negative values (default is 0.01).
    :return: Activated values.
    """
    return np.where(x > 0, x, x * alpha)

def leaky_relu_derivative(x, alpha=0.01):
    """
    Derivative of the Leaky ReLU function.
    :param x: Input data.
    :param alpha: Slope for negative values (default is 0.01).
    :return: Gradient of the Leaky ReLU function.
    """
    return np.where(x > 0, 1, alpha)
````

````python
class BaseLayer:
    def forward(self, inputs):
        raise NotImplementedError

    def backward(self, grad_output):
        raise NotImplementedError

class InputLayer(BaseLayer):
    def forward(self, inputs):
        return inputs

    def backward(self, grad_output):
        return grad_output

class DenseLayer(BaseLayer):

    def __init__(self, input_size, output_size):
        self.weights = np.random.randn(input_size, output_size) * 0.1
        self.biases = np.random.randn(1, output_size) * 0.1

    def forward(self, inputs):
        self.inputs = inputs
        return np.dot(inputs, self.weights) + self.biases

    def backward(self, grad_output, learning_rate):
        grad_input = np.dot(grad_output, self.weights.T)
        grad_weights = np.dot(self.inputs.T, grad_output)
        grad_biases = np.sum(grad_output, axis=0, keepdims=True)
        self.weights -= learning_rate * grad_weights
        self.biases -= learning_rate * grad_biases
        return grad_input

class ActivationLayer(BaseLayer):
    def __init__(self, activation, derivative):
        self.activation = activation
        self.derivative = derivative

    def forward(self, inputs):
        self.inputs = inputs
        self.output = self.activation(inputs)
        return self.output

    def backward(self, grad_output):
        if self.activation.__name__ == 'softmax':
            return np.einsum('ijk,ik->ij', softmax_derivative(self.output), grad_output)
        return grad_output * self.derivative(self.inputs)
````

````python
class ANN:
    def __init__(self):
        self.layers = []

    def add(self, layer):
        self.layers.append(layer)

    def forward(self, X):
        """
        Forward pass through all layers.
        :param X: Input data.
        :return: Network output.
        """
        output = X
        for layer in self.layers:
            output = layer.forward(output)
        return output

    def backward(self, grad_output, learning_rate, clip_value=1.0):
        """
        Backward pass through all layers in reverse order with gradient clipping.
        :param grad_output: Initial gradient.
        :param learning_rate: Learning rate for the updates.
        :param clip_value: Maximum allowed norm of the gradient.
        :return: Final gradient after backpropagation.
        """
        for layer in reversed(self.layers):
            if isinstance(layer, DenseLayer):
                # Compute gradient for layer
                grad_output = layer.backward(grad_output, learning_rate)
                # Clip the gradients to avoid explosion
                grad_norm = np.linalg.norm(grad_output)
                if grad_norm > clip_value:
                    grad_output = (grad_output / grad_norm) * clip_value
            else:
                grad_output = layer.backward(grad_output)
        return grad_output


    def train(self, X, y, epochs=1000, learning_rate=0.01, batch_size=32, clip_norm=1.0):

        num_samples = X.shape[1]
        for epoch in range(epochs):
            # Shuffle the data at the beginning of each epoch
            permutation = np.random.permutation(num_samples)
            X_shuffled = X[permutation, :]
            y_shuffled = y[permutation, :]

            for i in range(0, num_samples, batch_size):
                # Create mini-batch
                batch_end = min(i + batch_size, num_samples)
                X_batch = X_shuffled[i:batch_end, :]
                y_batch = y_shuffled[i:batch_end, :]

                # Forward pass
                output = self.forward(X_batch)

                output = np.clip(output, 1e-9, 1 - 1e-9)

                # Initial gradient (negative log-likelihood)
                grad_output = output - y_batch

                # Backward pass
                self.backward(grad_output, learning_rate, clip_norm)
                #gradients = clip_gradients(gradients, max_norm=clip_norm)

            if epoch % 50 == 0:
                final_output = self.forward(X)

                final_output = np.clip(final_output, 1e-9, 1 - 1e-9)
                # Compute loss (cross-entropy for classification)

                loss = np.mean(-np.log(np.sum(y * final_output, axis=1)))

                # Calculate accuracy
                acc = np.mean(np.argmax(y, axis=1) == np.argmax(final_output, axis=1))


                print(f"Epoch {epoch}, Loss: {loss:.4f}, Accuracy: {np.round(acc*100.0, 2)}%")

    def predict(self, X):
        return self.forward(X)
````
